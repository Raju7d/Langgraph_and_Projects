```python
import langgraph
```

```python
# we need to initialize the State: StateGraph

# ex: if you have idea, that idea want to give it to some one, you need to send some one

# what is my state? what is my current information

from langgraph.graph import StateGraph
```

```python
# 3 agents A--->B --> C

# everyone will one task
# everyone need info : state
# everyone will produce some info : state
```

```python
graph = StateGraph(dict) # info will provide using dict

# empty state
```

```python
def TaskA(state): # 
    print("A started working")
    return(state) # produce the state

def TaskB(state):
    print("B started working")
    return(state)

def TaskC(state):
    print("C started working")
    return(state)
```

```python
# add nodes
graph.add_node('A', TaskA)
graph.add_node('B', TaskB)
graph.add_node('C', TaskC)
```

```python
# add edges
graph.add_edge('A', 'B')
graph.add_edge('B', 'C')
```

```python
graph.set_entry_point('A')
```

```python
app = graph.compile()
```

# MCP (Model context protocol)

## MCP  flow

1. Whenever user gives an input, The **MCP client** send the request to the **MCP server**
2. MCP Server checks what all the **tools & information** available inside the **MCP Server**
3. Then **MCP Client (LLM)** sends the request to the **respective tool** to execute it
